```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import holoviews as hv
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

#### Using StandardScalar Method

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
df_market_data_scaled=StandardScaler().fit_transform(df_market_data)
df_market_data_scaled
```

```python
# Create a DataFrame with the scaled data
coin_scaled_df=pd.DataFrame(df_market_data_scaled,columns=["price_change_percentage_24h","price_change_percentage_7d","price_change_percentage_14d","price_change_percentage_30d","price_change_percentage_60d","price_change_percentage_200d","price_change_percentage_1y"])

# Copy the crypto names from the original data
coin_scaled_df["coin_id"]=df_market_data.index
# Set the coinid column as index
coin_scaled_df=coin_scaled_df.set_index("coin_id")

# Display sample data
coin_scaled_df.head()
```

---

### Find the Best Value for k Using the Original Data.

```python
# Create a list with the number of k-values from 1 to 11
k_values=list(range(1,11))
k_values
```

```python
# Create an empty list to store the inertia values
inertia = []

# Create a for loop to compute the inertia with each possible value of k
for i in k_values:
# Inside the loop:
    k_model=KMeans(n_clusters=i,random_state=1)
# 1. Create a KMeans model using the loop counter for the n_clusters
    k_model.fit(coin_scaled_df)
# 2. Fit the model to the data using `df_market_data_scaled`
    inertia.append(k_model.inertia_)
# 3. Append the model.inertia_ to the inertia list
```

```python
# Create a dictionary with the data to plot the Elbow curve
k_model_dict={"k": k_values,"inertia":inertia}

# Create a DataFrame with the data to plot the Elbow curve
k_model_df=pd.DataFrame(k_model_dict)
k_model_df
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
elbow_scaled=k_model_df.hvplot.line(
    x="k",
    y="inertia",
    xticks=k_values,
    title="Elbow Curve"
)
elbow_scaled
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** 4

### Clustering cryptocurrencies using Scaled Data

```python
# Model Data using K-Mean
model=KMeans(n_clusters=4,random_state=1)
```

```python
# Fit the Model
model.fit(coin_scaled_df)
```

```python
# Predict the Data based on model
predictions_scaled=model.predict(coin_scaled_df)
predictions_scaled
```

### Adding predictions to Original Data frame

```python
# Create a copy of the DataFrame
market_scaled_df=df_market_data.copy()
```

```python
# Add a new column to the DataFrame with the predicted clusters
market_scaled_df["cluster"]=predictions_scaled

# Display sample data
market_scaled_df.head(3)
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.
market_cluster=market_scaled_df.hvplot.scatter(
    x="price_change_percentage_24h",
    y="price_change_percentage_7d",
    by="cluster",
    hover_cols="coin_id",
    width=600,  
    height=400, 
    title="Market Scatter Plot with K-Means Clustering"
)
market_cluster
```

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca_3=PCA(n_components=3)
```

```python
# Use the PCA model with `fit_transform` to reduce to 
# three principal components.
model=pca_3.fit_transform(coin_scaled_df)
# View the first five rows of the DataFrame. 
model[0:5]
```

```python
# Retrieve the explained variance to determine how much information 
# can be attributed to each principal component.
explained_variance=pca_3.explained_variance_ratio_
explained_variance.sum()
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** 89

```python
# Create a new DataFrame with the PCA data.
coins_pca3_df=pd.DataFrame(model,columns=["PCA1","PCA2","PCA3"])
# Creating a DataFrame with the PCA data

coins_pca3_df["coin_id"]=df_market_data.index
# Set the coinid column as index
coins_pca3_df=coins_pca3_df.set_index("coin_id")

# Display sample data
coins_pca3_df.head()
```

---

### Find the Best Value for k Using the PCA Data

```python
# Create a list with the number of k-values from 1 to 11
k_pca3=list(range(1,11))
k_pca3
```

```python
# Create an empty list to store the inertia values
interia_pca3=[]

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
for i in k_pca3:
# 1. Create a KMeans model using the loop counter for the n_clusters
    model_pca3=KMeans(n_clusters=i,random_state=1)
# 2. Fit the model to the data using `df_market_data_pca`
    model_pca3.fit(coins_pca3_df)
# 3. Append the model.inertia_ to the inertia list
    interia_pca3.append(model_pca3.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
pca3_dict={"k_pca":k_pca3,"inertia_pca":interia_pca3}
# Create a DataFrame with the data to plot the Elbow curve
pca3_df=pd.DataFrame(pca3_dict)
pca3_df
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
elbow_pca=pca3_df.hvplot.line(
    x="k_pca",
    y="inertia_pca",
    xticks=k_pca3
)
elbow_pca
```

#### Answer the following questions: 

* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer:** 4


* **Question:** Does it differ from the best k value found using the original data?

  * **Answer:** No, original data has 4 value for cluster and PCA data suggest 4 cluster value.

### Cluster Cryptocurrencies with K-means Using the PCA Data

```python
# Initialize the K-Means model using the best value for k
k_model_pca=KMeans(n_clusters=4,random_state=1)
```

```python
# Fit the K-Means model using the PCA data
k_model_pca.fit(coins_pca3_df)
```

```python
# Predict the clusters to group the cryptocurrencies using the PCA data
predictions_pca=k_model_pca.predict(coins_pca3_df)
# Print the resulting array of cluster values.
predictions_pca
```

```python
# Create a copy of the DataFrame with the PCA data
pca_data_coins=coins_pca3_df.copy()

# Add a new column to the DataFrame with the predicted clusters
pca_data_coins["pca_cluster"]=predictions_pca

# Display sample data
pca_data_coins
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="PC1"` and `y="PC2"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.
pca_cluster=pca_data_coins.hvplot.scatter(
    x="PCA1",
    y="PCA2",
    by="pca_cluster",
    hover_cols=["coin_id"],
    width=600,  
    height=400, 
    title="PCA Scatter Plot with K-Means Clustering"
)
pca_cluster
```
